## <font color=73b2ff size=3>3.2.</font> **Operações e Controle de Dados**

Nesta subseção estão reunidas as rotinas de organização das planilhas, visualização por abas, formulário de atualização de dados e controle de acesso ao painel administrativo. O objetivo é manter, em um único bloco, as funções operacionais centrais do notebook.

```python
# --- Constantes do Projeto ---
PASTA_CONFIGURACOES = "DadosEmPlanilhas"  # Nome da pasta onde os arquivos de configuração serão armazenados

# --- Funções Auxiliares ---
def verificar_e_criar_pasta_config(caminho_pasta: str) -> bool:
    """
    Verifica se a pasta de configurações existe e a cria se necessário.

    Args:
        caminho_pasta (str): O caminho para a pasta de configurações.

    Returns:
        bool: True se a pasta existe ou foi criada com sucesso, False caso contrário.
    """
    if not os.path.exists(caminho_pasta):
        try:
            os.makedirs(caminho_pasta)
            print(f"[INFO] Pasta de configurações '{caminho_pasta}' criada com sucesso.")
            return True
        except OSError as e:
            print(f"[ERRO] Não foi possível criar a pasta '{caminho_pasta}': {e}")
            return False

    print(f"[INFO] Pasta de configurações '{caminho_pasta}' já existe.")
    return True


def _validar_e_processar_config_aba(config_linha_dict: dict) -> dict | None:
    """
    Valida e processa uma linha de configuração de uma aba.
    Converte tipos de dados e formata listas de colunas.

    Args:
        config_linha_dict (dict): Dicionário com as configurações de uma única aba.

    Returns:
        dict | None: Dicionário de configuração validado e processado, ou None se inválido.
    """
    campos_obrigatorios = [
        "nome_arquivo_principal",
        "nome_aba_principal",
        "linha_cabecalho",
        "linha_inicio_dados",
        "tipo_dado"
    ]

    for campo in campos_obrigatorios:
        if campo not in config_linha_dict or pd.isna(config_linha_dict[campo]):
            print(f"[ALERTA] Configuração ignorada: Campo obrigatório '{campo}' ausente ou vazio.")
            return None

    config_processada = config_linha_dict.copy()

    try:
        config_processada["linha_cabecalho"] = int(config_processada["linha_cabecalho"])
        config_processada["linha_inicio_dados"] = int(config_processada["linha_inicio_dados"])

        if "colunas_utilizadas" in config_processada and pd.notna(config_processada["colunas_utilizadas"]):
            config_processada["colunas_utilizadas"] = [
                col.strip() for col in str(config_processada["colunas_utilizadas"]).split(",") if col.strip()
            ]
        else:
            config_processada["colunas_utilizadas"] = []

        if "colunas_ignorar" in config_processada and pd.notna(config_processada["colunas_ignorar"]):
            config_processada["colunas_ignorar"] = [
                col.strip() for col in str(config_processada["colunas_ignorar"]).split(",") if col.strip()
            ]
        else:
            config_processada["colunas_ignorar"] = []

        return config_processada

    except ValueError as e:
        print(
            f"[ERRO] Erro de tipo na configuração para "
            f"'{config_linha_dict.get('nome_arquivo_principal', 'N/A')}' - "
            f"'{config_linha_dict.get('nome_aba_principal', 'N/A')}': {e}"
        )
        return None


def ler_arquivo_configuracao(caminho_arquivo_config: str) -> dict:
    """
    Lê um arquivo Excel de configuração e extrai as regras para as abas.

    Args:
        caminho_arquivo_config (str): O caminho completo para o arquivo Excel de configuração.

    Returns:
        dict: Dicionário contendo as configurações organizadas por
              {nome_arquivo_principal: {nome_aba_principal: {detalhes_config}}}.
    """
    configs_do_arquivo = {}

    try:
        # Suprime o UserWarning específico do openpyxl sobre 'Print area'
        with warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl.reader.workbook'):
            df_config = pd.read_excel(caminho_arquivo_config)

        if df_config.empty:
            print(f"[ALERTA] O arquivo de configuração '{os.path.basename(caminho_arquivo_config)}' está vazio.")
            return configs_do_arquivo

        for _, row in df_config.iterrows():
            config_linha_dict = row.to_dict()
            config_processada = _validar_e_processar_config_aba(config_linha_dict)

            if config_processada:
                nome_arquivo_principal = config_processada["nome_arquivo_principal"]
                nome_aba_principal = config_processada["nome_aba_principal"]

                if nome_arquivo_principal not in configs_do_arquivo:
                    configs_do_arquivo[nome_arquivo_principal] = {}

                configs_do_arquivo[nome_arquivo_principal][nome_aba_principal] = config_processada

    except FileNotFoundError:
        print(f"[ERRO] Arquivo de configuração não encontrado: '{caminho_arquivo_config}'.")
    except Exception as e:
        print(f"[ERRO] Erro ao ler o arquivo de configuração '{caminho_arquivo_config}': {e}")

    return configs_do_arquivo


def carregar_todas_configuracoes(caminho_pasta_config: str) -> dict:
    """
    Carrega todas as configurações de todos os arquivos Excel na pasta de configurações.

    Args:
        caminho_pasta_config (str): O caminho para a pasta contendo os arquivos Excel de configuração.

    Returns:
        dict: Um dicionário consolidado com todas as configurações encontradas.
    """
    configuracoes_globais = {}

    if not verificar_e_criar_pasta_config(caminho_pasta_config):
        return configuracoes_globais

    print(f"[INFO] Buscando arquivos de configuração na pasta: '{caminho_pasta_config}'")

    for filename in os.listdir(caminho_pasta_config):
        if filename.endswith((".xlsx", ".xls")):
            file_path = os.path.join(caminho_pasta_config, filename)
            print(f"[INFO] Processando arquivo: '{filename}'")

            configs_do_arquivo = ler_arquivo_configuracao(file_path)

            for nome_arquivo_principal, abas_config in configs_do_arquivo.items():
                if nome_arquivo_principal not in configuracoes_globais:
                    configuracoes_globais[nome_arquivo_principal] = {}

                configuracoes_globais[nome_arquivo_principal].update(abas_config)
        else:
            print(f"[INFO] Ignorando arquivo não Excel: '{filename}'")

    return configuracoes_globais


def exibir_resumo_configuracoes(configuracoes: dict):
    """
    Exibe um resumo formatado das configurações de leitura carregadas.

    Args:
        configuracoes (dict): Dicionário contendo todas as configurações carregadas.
    """
    print("\n--- Resumo das Configurações de Leitura Carregadas ---")

    if not configuracoes:
        print("Nenhuma configuração de planilha foi carregada.")
        return

    for nome_arquivo, abas_config in configuracoes.items():
        print(f"\n### Planilha Principal: '{nome_arquivo}' ###")

        for nome_aba, detalhes in abas_config.items():
            print(f"  - Aba: '{nome_aba}' (Tipo de Dado: {detalhes.get('tipo_dado', 'N/A')})")
            print(f"    - Cabeçalho na linha: {detalhes.get('linha_cabecalho', 'N/A')}")
            print(f"    - Dados iniciam na linha: {detalhes.get('linha_inicio_dados', 'N/A')}")
            print(f"    - Colunas utilizadas: {', '.join(detalhes.get('colunas_utilizadas', ['Todas']))}")
            print(f"    - Colunas ignoradas: {', '.join(detalhes.get('colunas_ignorar', ['Nenhuma']))}")
            print(f"    - Observação: {detalhes.get('observacao', 'N/A')}")

    print("\n------------------------------------------------------")
```

```python
def exibir_planilhas_por_abas(caminho_pasta: str):
    """
    Exibe o conteúdo de cada aba dos arquivos Excel contidos na pasta informada,
    organizando a visualização por arquivo e por aba.
    """
    if not os.path.exists(caminho_pasta):
        print(f"A pasta '{caminho_pasta}' não foi encontrada.")
        return

    print(f"\n--- Conteúdo das planilhas na pasta '{caminho_pasta}' ---\n")
    encontrou_arquivo = False

    for filename in os.listdir(caminho_pasta):
        if filename.endswith((".xlsx", ".xls")):
            encontrou_arquivo = True
            file_path = os.path.join(caminho_pasta, filename)
            print(f"\n### Arquivo: {filename} ###")

            try:
                xls = pd.ExcelFile(file_path)
                sheet_names = xls.sheet_names

                if not sheet_names:
                    print(f"  [ALERTA] O arquivo '{filename}' não contém abas.\n")
                    continue

                for sheet_name in sheet_names:
                    print(f"  #### Aba: {sheet_name} ####")
                    df = pd.read_excel(xls, sheet_name=sheet_name)

                    if df.empty:
                        print("    Aba vazia.")
                    else:
                        display(df)

                    print("\n")

            except Exception as e:
                print(f"  [ERRO] Não foi possível ler o arquivo '{filename}' ou uma de suas abas: {e}\n")

    if not encontrou_arquivo:
        print("Nenhum arquivo Excel encontrado na pasta.")

    print("\n------------------------------------------------------")
```

#### * Leitura Estruturada das Planilhas

```python
# -------------------------------------------------------
# Leitura Estruturada das Planilhas
# -------------------------------------------------------

def ler_planilha_configurada(caminho_arquivo: str, config_aba: dict) -> pd.DataFrame | None:
    try:
        df = pd.read_excel(
            caminho_arquivo,
            sheet_name=config_aba["nome_aba_principal"],
            header=config_aba["linha_cabecalho"] - 1
        )

        # Remove linhas anteriores ao início dos dados, se necessário
        linha_inicio = config_aba["linha_inicio_dados"] - 1
        if linha_inicio > 0:
            df = df.iloc[linha_inicio:].reset_index(drop=True)

        # Seleciona colunas específicas, se houver
        colunas_utilizadas = config_aba.get("colunas_utilizadas", [])
        colunas_ignorar = config_aba.get("colunas_ignorar", [])

        if colunas_utilizadas:
            colunas_existentes = [col for col in colunas_utilizadas if col in df.columns]
            df = df[colunas_existentes]

        if colunas_ignorar:
            df = df.drop(columns=colunas_ignorar, errors="ignore")

        return df

    except Exception as e:
        print(f"Erro ao ler a aba '{config_aba.get('nome_aba_principal', 'N/A')}': {e}")
        return None
```

```python
# -------------------------------------------------------
# Carregamento Geral das Bases Configuradas
# -------------------------------------------------------

def carregar_bases_configuradas(configuracoes: dict, pasta_base: str) -> dict:
    bases_carregadas = {}

    for nome_arquivo, abas_config in configuracoes.items():
        caminho_arquivo = os.path.join(pasta_base, nome_arquivo)

        if not os.path.exists(caminho_arquivo):
            print(f"Arquivo principal não encontrado: {caminho_arquivo}")
            continue

        bases_carregadas[nome_arquivo] = {}

        for nome_aba, config_aba in abas_config.items():
            df = ler_planilha_configurada(caminho_arquivo, config_aba)

            if df is not None:
                bases_carregadas[nome_arquivo][nome_aba] = df
                print(f"Base carregada com sucesso: {nome_arquivo} -> {nome_aba}")

    return bases_carregadas
```

#### * Tratamento Inicial dos Dados

```python
# -------------------------------------------------------
# Tratamento Inicial dos Dados
# -------------------------------------------------------

def padronizar_colunas(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [
        str(col).strip().lower().replace(" ", "_")
        for col in df.columns
    ]
    return df

def remover_linhas_vazias(df: pd.DataFrame) -> pd.DataFrame:
    return df.dropna(how="all").reset_index(drop=True)

def tratamento_inicial(df: pd.DataFrame) -> pd.DataFrame:
    df = remover_linhas_vazias(df)
    df = padronizar_colunas(df)
    return df
```

#### * Gerar Configuração de IES e Programas

```python
def gerar_rede_ies_programas_profissional(df_programas, df_bolsistas=None, ocultar_programas=None):
    from google.colab import files

    if df_programas is None or df_programas.empty:
        print("Erro: base de programas vazia.")
        return

    df = df_programas.copy()

    # Padronização inicial
    for col in ["programa_de_fomento", "instituicao", "tipo_ies"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip().str.upper()

    # FILTRO DE PROGRAMAS
    if ocultar_programas:
        # Converte a lista do usuário para maiúsculas para bater com a padronização
        lista_filtro = [str(p).strip().upper() for p in ocultar_programas]

        # EXPLICAÇÃO PARA INVERTER:
        # Para OCULTAR os programas da lista, usamos o til (~) antes do df['...'].
        # Para EXIBIR APENAS os programas da lista, remova o caractere ~ abaixo.
        df = df[~df['programa_de_fomento'].isin(lista_filtro)]

    if "id" not in df.columns:
        df["id"] = df.index.astype(str)

    if "tipo_ies" not in df.columns:
        df["tipo_ies"] = "CIVIL"

    df["qntd_de_bolsistas"] = pd.to_numeric(df.get("qntd_de_bolsistas", 0), errors="coerce").fillna(0)

    # Lógica de pesos (bolsistas)
    if df_bolsistas is not None and not df_bolsistas.empty and "id" in df_bolsistas.columns:
        df_b = df_bolsistas.copy()
        df_b["id"] = df_b["id"].astype(str).str.strip()
        df["id"] = df["id"].astype(str).str.strip()
        contagem_bolsistas = df_b.groupby("id").size().reset_index(name="bolsistas_detalhado")
        df = df.merge(contagem_bolsistas, on="id", how="left")
        df["bolsistas_detalhado"] = df["bolsistas_detalhado"].fillna(0)
        df["peso_final"] = df[["qntd_de_bolsistas", "bolsistas_detalhado"]].max(axis=1)
    else:
        df["peso_final"] = df["qntd_de_bolsistas"]

    df = df.dropna(subset=["programa_de_fomento", "instituicao"])

    # Agrupamentos para o Grafo
    arestas = df.groupby(["programa_de_fomento", "instituicao"], as_index=False).agg(
        total_projetos=("id", "nunique"),
        total_bolsistas=("peso_final", "sum"),
        tipo_ies=("tipo_ies", "first")
    )

    # --- FILTRO: Não exibir IES com 0 bolsistas ---
    arestas = arestas[arestas["total_bolsistas"] > 0]

    programas = arestas.groupby("programa_de_fomento", as_index=False).agg(
        projetos=("total_projetos", "sum"),
        bolsistas=("total_bolsistas", "sum")
    )

    ies = arestas.groupby("instituicao", as_index=False).agg(
        projetos=("total_projetos", "sum"),
        bolsistas=("total_bolsistas", "sum"),
        tipo_ies=("tipo_ies", "first")
    )

    if arestas.empty:
        print("Aviso: Após os filtros, não restaram dados com bolsistas para gerar a rede.")
        return

    def escalar(valor, vmin, vmax, minimo, maximo):
        if vmax == vmin:
            return (minimo + maximo) / 2
        return minimo + ((valor - vmin) / (vmax - vmin)) * (maximo - minimo)

    G = nx.Graph()

    # Adição de Nós (Programas)
    pmin, pmax = programas["projetos"].min(), programas["projetos"].max()
    for _, row in programas.iterrows():
        tamanho = escalar(row["projetos"], pmin, pmax, 35, 70)
        label_prog = f"{row['programa_de_fomento']}\n {int(row['projetos'])} Projeto(s)"
        G.add_node(
            row["programa_de_fomento"],
            label=label_prog,
            size=tamanho,
            color="#003366",
            shape="dot",
            title=f"Programa:{row['programa_de_fomento']}\nProjetos: {int(row['projetos'])}\nBolsistas: {int(row['bolsistas'])}"
        )

    # Adição de Nós (IES)
    imin, imax = ies["projetos"].min(), ies["projetos"].max()
    for _, row in ies.iterrows():
        tamanho = escalar(row["projetos"], imin, imax, 18, 45)
        tipo = str(row["tipo_ies"]).upper()
        cor = "#006400" if tipo in ["EB", "MB", "FAB", "MD", "MILITAR"] else "#28A745"
        label_ies = f"{row['instituicao']}\n {int(row['bolsistas'])} Bolsista(s)"
        G.add_node(
            row["instituicao"],
            label=label_ies,
            size=tamanho,
            color=cor,
            shape="dot",
            title=f"IES: {row['instituicao']}\nTipo: {tipo}\nProjetos: {int(row['projetos'])}\nBolsistas: {int(row['bolsistas'])}"
        )

    # Adição de Arestas
    bmin, bmax = arestas["total_bolsistas"].min(), arestas["total_bolsistas"].max()
    for _, row in arestas.iterrows():
        largura = escalar(row["total_bolsistas"], bmin, bmax, 1.5, 8)
        G.add_edge(
            row["programa_de_fomento"],
            row["instituicao"],
            width=largura,
            color="#C0C0C0",
            title=f"Projetos: {int(row['total_projetos'])}\nBolsistas:{int(row['total_bolsistas'])}"
        )

    os.makedirs("Analises_Rede_HTML", exist_ok=True)
    caminho_html = "Analises_Rede_HTML/Rede_Institucional_Profissional.html"

    net = Network(notebook=False, width="100%", height="750px", bgcolor="white", font_color="black", cdn_resources="remote")
    net.from_nx(G)
    net.force_atlas_2based()
    net.write_html(caminho_html)

    print(f"Rede gerada com sucesso em: {caminho_html}")
    display(HTML(filename=caminho_html))
    files.download(caminho_html)
```

#### * Conexão Parceiros <-> Programas de Fomento <-> IES

```python
# ---
# Rede Hierárquica: Parceiros (CAPES/CNPq) <-> Programas <-> IES
# ---
def gerar_rede_parceiros_prog_ies_profissional(df_programas, df_bolsistas=None, ocultar_programas=None):
    from google.colab import files

    if df_programas is None or df_programas.empty:
        print("Erro: base de programas vazia.")
        return

    df = df_programas.copy()

    # Padronização inicial
    for col in ["programa_de_fomento", "instituicao", "tipo_ies"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip().str.upper()

    # Filtro de Programas
    if ocultar_programas:
        lista_filtro = [str(p).strip().upper() for p in ocultar_programas]
        df = df[df['programa_de_fomento'].isin(lista_filtro)]

    # Função para identificar o parceiro
    def identificar_parceiro(programa):
        prog = str(programa).upper()
        if any(p in prog for p in ["PRÓ-DEFESA", "PRO-DEFESA", "PROCAD-DEFESA", "PROPEX-DEFESA", "ÁLVARO ALBERTO"]):
            return "CAPES"
        return "CNPQ"

    if "id" not in df.columns:
        df["id"] = df.index.astype(str)

    df["qntd_de_bolsistas"] = pd.to_numeric(df.get("qntd_de_bolsistas", 0), errors="coerce").fillna(0)

    # Peso final (considerando base de bolsistas se disponível)
    if df_bolsistas is not None and not df_bolsistas.empty and "id" in df_bolsistas.columns:
        df_b = df_bolsistas.copy()
        df_b["id"] = df_b["id"].astype(str).str.strip()
        df["id"] = df["id"].astype(str).str.strip()
        contagem = df_b.groupby("id").size().reset_index(name="bolsistas_det")
        df = df.merge(contagem, on="id", how="left").fillna(0)
        df["peso_final"] = df[["qntd_de_bolsistas", "bolsistas_det"]].max(axis=1)
    else:
        df["peso_final"] = df["qntd_de_bolsistas"]

    G = nx.Graph()

    # Agrupamentos
    prog_ies = df.groupby(["programa_de_fomento", "instituicao"], as_index=False).agg(bolsistas=("peso_final", "sum"), tipo_ies=("tipo_ies", "first"))
    programas = df.groupby("programa_de_fomento", as_index=False).agg(projetos=("id", "nunique"))
    ies_lista = df.groupby("instituicao", as_index=False).agg(bolsistas=("peso_final", "sum"), tipo_ies=("tipo_ies", "first"))

    # 1. Adicionar Parceiros
    for p in ["CAPES", "CNPQ"]:
        G.add_node(p, label=p, size=60, color="#FFA500", shape="diamond", title=f"Parceiro: {p}")

    # 2. Adicionar Programas e ligar aos Parceiros
    for _, row in programas.iterrows():
        prog = row["programa_de_fomento"]
        parceiro = identificar_parceiro(prog)
        G.add_node(prog, label=f"{prog}\n{int(row['projetos'])} Projeto(s)", size=40, color="#003366", shape="dot")
        G.add_edge(parceiro, prog, width=2, color="#888888")

    # 3. Adicionar IES e ligar aos Programas
    for _, row in prog_ies.iterrows():
        ies_nome = row["instituicao"]
        prog_nome = row["programa_de_fomento"]
        tipo = str(row["tipo_ies"]).upper()
        cor_ies = "#006400" if tipo in ["EB", "MB", "FAB", "MD", "MILITAR"] else "#28A745"

        if ies_nome not in G:
            total_bolsistas_ies = ies_lista[ies_lista["instituicao"] == ies_nome]["bolsistas"].values[0]
            G.add_node(ies_nome, label=f"{ies_nome}\n{int(total_bolsistas_ies)} Bolsista(s)", size=25, color=cor_ies, shape="dot")

        G.add_edge(prog_nome, ies_nome, width=1.5, color="#C0C0C0", title=f"Bolsistas: {int(row['bolsistas'])}")

    os.makedirs("Analises_Rede_HTML", exist_ok=True)
    caminho_html = "Analises_Rede_HTML/rede_parceiro_programa_ies_profissional.html"
    net = Network(notebook=False, width="100%", height="800px", bgcolor="white", font_color="black", cdn_resources="remote")
    net.from_nx(G)
    net.force_atlas_2based()
    net.write_html(caminho_html)

    print(f"Rede gerada com sucesso em: {caminho_html}")
    display(HTML(filename=caminho_html))
    files.download(caminho_html)
```

#### * Nacional - Conexão IES <-> Programas de Fomento

```python
# -------------------------------------------------------------------
# Seção 4: Análise de Redes - Conexão IES <-> Programas (Brasil)
# -------------------------------------------------------------------
def gerar_rede_nacional(df_programas):
    """
    Gera um grafo interativo conectando Programas de Fomento às IES.
    Filtra apenas registros onde o país é Brasil.
    """

      # --- CRIAÇÃO DA PASTA ---
    PASTA_SAIDA = "Analises_Rede_HTML"
    if not os.path.exists(PASTA_SAIDA): os.makedirs(PASTA_SAIDA)

    # --- CONFIGURAÇÕES DE ESTILO ---
    COR_PROGRAMA = "blue"
    COR_IES = "red"
    COR_FUNDO = "#222222"
    COR_FONTE = "white"

    # --- VALIDAÇÃO ---
    if df_programas is None or df_programas.empty:
        print("Erro: DataFrame vazio ou não fornecido.")
        return

    required_cols = ["programa_de_fomento", "instituicao", "qntd_de_bolsistas", "pais"]
    missing_cols = [col for col in required_cols if col not in df_programas.columns]
    if missing_cols:
        print(f"Erro: Colunas necessárias não encontradas: {', '.join(missing_cols)}")
        return

    # --- PREPARAÇÃO DOS DADOS (FILTRO NACIONAL) ---
    df = df_programas.copy()
    df["pais"] = df["pais"].astype(str).str.strip().str.lower()
    lista_brasil = ["brasil", "brazil", "brasileiro", "brasileira"]
    df = df[df["pais"].isin(lista_brasil)]

    if df.empty:
        print("Nenhum dado encontrado para o filtro 'Brasil'.")
        return

    df = df.dropna(subset=["programa_de_fomento", "instituicao"])
    df["qntd_de_bolsistas"] = pd.to_numeric(df["qntd_de_bolsistas"], errors='coerce').fillna(0)
    df_grouped = df.groupby(["programa_de_fomento", "instituicao"])[["qntd_de_bolsistas"]].sum().reset_index()

    # --- CONSTRUÇÃO DO GRAFO ---
    G = nx.Graph()

    for _, row in df_grouped.iterrows():
        programa = str(row['programa_de_fomento']).strip().upper()
        ies = str(row['instituicao']).strip().upper()
        bolsistas_count = int(row['qntd_de_bolsistas'])

        if bolsistas_count == 0 or programa in ["", "NAN"] or ies in ["", "NAN"]:
            continue

        G.add_node(programa, group='Programa', color=COR_PROGRAMA, size=40, title=f"Programa: {programa}")
        G.add_node(ies, group='IES', color=COR_IES, size=30, title=f"IES: {ies}")
        G.add_edge(programa, ies, value=bolsistas_count, title=f"Bolsistas: {bolsistas_count}", weight=bolsistas_count)

    if G.number_of_nodes() == 0:
        print("Aviso: Nenhum nó foi adicionado à rede após o filtro.")
        return

    print(f"Gerando rede NACIONAL (Brasil) com {G.number_of_nodes()} nós e {G.number_of_edges()} arestas.")

    # --- VISUALIZAÇÃO ---
    net = Network(notebook=True, width="100%", height="800px", bgcolor=COR_FUNDO, font_color=COR_FONTE, cdn_resources='remote', directed=False)
    net.from_nx(G)

    net.set_options("""
    {
      "edges": {
        "color": {"inherit": true},
        "smooth": {"enabled": true, "type": "continuous"},
        "scaling": {"min": 1, "max": 12}
      },
      "physics": {
        "barnesHut": {
          "gravitationalConstant": -8000,
          "centralGravity": 0.3,
          "springLength": 250,
          "springConstant": 0.05
        },
        "minVelocity": 0.75
      }
    }
    """)

    if G.number_of_nodes() > 200:
        net.toggle_physics(False)

    caminho_arquivo = os.path.join(PASTA_SAIDA, "rede_nacional_brasil.html")
    net.show(caminho_arquivo)
    display(HTML(caminho_arquivo))
```

#### * Conexões de IES sobre Projetos

```python
# -----------------------------------------------------------------------
# Seção 6: Análise de Redes - Conexão IES Líder (Programas) <-> IES Participante (Bolsistas)
# -----------------------------------------------------------------------
def gerar_rede_ies_lid_bols(df_programas, df_bolsistas):
    """
    Gera um grafo baseado no cruzamento de IDs.
    O peso da aresta e o tamanho dos nós refletem a contagem de bolsistas.
    """
    from google.colab import files

    PASTA_SAIDA = "Analises_Rede_HTML"
    if not os.path.exists(PASTA_SAIDA): os.makedirs(PASTA_SAIDA)

    if df_programas is None or df_bolsistas is None:
        print("Erro: Bases de dados não fornecidas.")
        return

    # Normalização de IDs para garantir o merge
    df_p = df_programas[['id', 'instituicao']].copy()
    df_p['id'] = df_p['id'].astype(str).str.strip()

    df_b = df_bolsistas[['id', 'sigla_ies', 'tipo_ies']].copy() if 'tipo_ies' in df_bolsistas.columns else df_bolsistas[['id', 'sigla_ies']].copy()
    df_b['id'] = df_b['id'].astype(str).str.strip()

    # Cruzamento de dados
    df_merge = pd.merge(df_b, df_p, on='id', how='inner')

    if df_merge.empty:
        print("Aviso: Nenhum cruzamento encontrado entre programas e bolsistas.")
        return

    # Lógica de contagem
    df_links = df_merge.groupby(['instituicao', 'sigla_ies']).size().reset_index(name='quantidade')

    # Tenta obter o tipo_ies se disponível
    if 'tipo_ies' in df_merge.columns:
        dict_tipos = df_merge.set_index('sigla_ies')['tipo_ies'].to_dict()
    else:
        dict_tipos = {}

    def escalar(valor, vmin, vmax, minimo, maximo):
        if vmax == vmin: return (minimo + maximo) / 2
        return minimo + ((valor - vmin) / (vmax - vmin)) * (maximo - minimo)

    G = nx.Graph()

    contagem_lider = df_links.groupby('instituicao')['quantidade'].sum()
    contagem_part = df_links.groupby('sigla_ies')['quantidade'].sum()

    lmin, lmax = contagem_lider.min(), contagem_lider.max()
    pmin, pmax = contagem_part.min(), contagem_part.max()

    for _, row in df_links.iterrows():
        lider = str(row['instituicao']).strip().upper()
        part = str(row['sigla_ies']).strip().upper()
        peso = int(row['quantidade'])

        # Nodo Líder
        total_lider = int(contagem_lider[row['instituicao']])
        size_lider = escalar(total_lider, lmin, lmax, 30, 60)
        G.add_node(lider, label=f"{lider}\n {total_lider} Bolsista(s)", color="#0330ab", size=size_lider, shape="dot",
                   title=f"IES Líder: {lider}\nTotal Bolsistas: {total_lider}")

        # Nodo Participante
        total_part = int(contagem_part[row['sigla_ies']])
        size_part = escalar(total_part, pmin, pmax, 15, 40)
        tipo = str(dict_tipos.get(row['sigla_ies'], "")).upper()
        cor_part = "darkgreen" if tipo in ["EB", "MB", "FAB", "MD"] else "#4073ff"

        G.add_node(part, label=f"{part}\n {total_part} Bolsista(s)", color=cor_part, size=size_part, shape="dot",
                   title=f"IES Participante: {part}\nBolsistas: {total_part}")

        G.add_edge(lider, part, value=peso, title=f"{peso} Bolsista(s)", color="#888888", width=1 + (peso * 0.2))

    net = Network(notebook=False, width="100%", height="800px", bgcolor="white", font_color="black", cdn_resources='remote')
    net.from_nx(G)
    net.force_atlas_2based()

    caminho_arquivo = os.path.join(PASTA_SAIDA, "rede_ies_lider_bolsistas.html")
    net.write_html(caminho_arquivo)

    print(f"Rede gerada com sucesso em: {caminho_arquivo}")
    display(HTML(filename=caminho_arquivo))
    files.download(caminho_arquivo)
```

#### * Gerar Pasta de Dados Consolidados

```python
# -------------------------------------------------------
# Pasta padrão para leitura dos Dados Consolidados
# -------------------------------------------------------

PASTA_DADOS_CONSOLIDADOS = "/content/DadosConsolidados"

def obter_arquivo_dados_consolidados():
    """
    Procura automaticamente um arquivo Excel dentro da pasta padrão
    de Dados Consolidados. Caso a pasta não exista, tenta criá-la.
    """
    try:
        if not os.path.exists(PASTA_DADOS_CONSOLIDADOS):
            os.makedirs(PASTA_DADOS_CONSOLIDADOS)
            print(f"Pasta '{PASTA_DADOS_CONSOLIDADOS}' criada com sucesso.")
            print("Insira o arquivo Excel de Dados Consolidados dentro dessa pasta.")
            return None
    except Exception as e:
        print(f"Erro ao criar/verificar a pasta '{PASTA_DADOS_CONSOLIDADOS}': {e}")
        return None

    try:
        arquivos_excel = [
            arq for arq in os.listdir(PASTA_DADOS_CONSOLIDADOS)
            if arq.endswith((".xlsx", ".xls"))
        ]

        if not arquivos_excel:
            print(f"Nenhum arquivo Excel foi encontrado na pasta '{PASTA_DADOS_CONSOLIDADOS}'.")
            return None

        if len(arquivos_excel) > 1:
            print(f"Foram encontrados múltiplos arquivos na pasta '{PASTA_DADOS_CONSOLIDADOS}'.")
            print(f"Será utilizado o primeiro arquivo encontrado: {arquivos_excel[0]}")

        caminho_arquivo = os.path.join(PASTA_DADOS_CONSOLIDADOS, arquivos_excel[0])
        print(f"Arquivo de Dados Consolidados identificado: {arquivos_excel[0]}")
        return caminho_arquivo

    except Exception as e:
        print(f"Erro ao localizar o arquivo de Dados Consolidados: {e}")
        return None
```

#### * Leitura e Processamento da planilha

```python
# -------------------------------------------------------
# Leitura e processamento da planilha de Dados Consolidados
# -------------------------------------------------------


def normalizar_texto(texto):
    return (
        str(texto)
        .strip()
        .lower()
        .replace(" ", "_")
        .replace("-", "_")
        .replace("/", "_")
        .replace("ç", "c")
        .replace("ã", "a")
        .replace("á", "a")
        .replace("à", "a")
        .replace("é", "e")
        .replace("ê", "e")
        .replace("í", "i")
        .replace("ó", "o")
        .replace("õ", "o")
        .replace("ô", "o")
        .replace("ú", "u")
        .replace("(", "")
        .replace(")", "")
    )


def padronizar_colunas(df):
    df = df.copy()
    df.columns = [normalizar_texto(col) for col in df.columns]
    return df


def obter_arquivo_dados_consolidados():
    """
    Procura automaticamente um arquivo Excel dentro da pasta padrão.
    """
    if not os.path.exists(PASTA_DADOS_CONSOLIDADOS):
        os.makedirs(PASTA_DADOS_CONSOLIDADOS)
        print(f"Pasta '{PASTA_DADOS_CONSOLIDADOS}' criada. Insira o arquivo Excel nela.")
        return None

    arquivos_excel = [
        arq for arq in os.listdir(PASTA_DADOS_CONSOLIDADOS)
        if arq.endswith((".xlsx", ".xls"))
    ]

    if not arquivos_excel:
        print(f"Nenhum arquivo Excel foi encontrado na pasta '{PASTA_DADOS_CONSOLIDADOS}'.")
        return None

    if len(arquivos_excel) > 1:
        print("Foram encontrados múltiplos arquivos na pasta.")
        print(f"Será utilizado o primeiro arquivo encontrado: {arquivos_excel[0]}")

    caminho_arquivo = os.path.join(PASTA_DADOS_CONSOLIDADOS, arquivos_excel[0])
    print(f"Arquivo identificado: {arquivos_excel[0]}")
    return caminho_arquivo


def identificar_tipo_aba(df):
    """
    Identifica o tipo lógico da aba com base nas colunas.
    """
    colunas = set(df.columns)

    if {"beneficiario_bolsista", "modalidade", "ies_extenso", "uf"}.issubset(colunas):
        return "bolsistas"

    if {"nome", "funcao_atuacao", "ies_extenso", "uf"}.issubset(colunas):
        return "nao_bolsistas"

    if {"programa_de_fomento", "titulo_do_projeto", "instituicao", "ies_extenso", "pais"}.issubset(colunas):
        return "programas_fomento"

    if {"ies_siglas", "nome_extenso"}.issubset(colunas):
        return "ies_extenso"

    return "outra"


def carregar_abas_relevantes():
    """
    Carrega e classifica as abas relevantes da planilha consolidada.
    """
    caminho_arquivo = obter_arquivo_dados_consolidados()

    if caminho_arquivo is None:
        return {}

    abas_carregadas = {}

    try:
        xls = pd.ExcelFile(caminho_arquivo)

        for aba in xls.sheet_names:
            try:
                df = pd.read_excel(caminho_arquivo, sheet_name=aba)
                df = padronizar_colunas(df)

                if df.empty:
                    continue

                tipo_aba = identificar_tipo_aba(df)

                if tipo_aba != "outra":
                    abas_carregadas[tipo_aba] = df
                    print(f"Aba '{aba}' identificada como '{tipo_aba}'.")
                    clear_alert()

            except Exception as e:
                print(f"Erro ao ler a aba '{aba}': {e}")

    except Exception as e:
        print(f"Erro ao abrir a planilha consolidada: {e}")

    return abas_carregadas
```

#### * Funções para a análise

```python
# -------------------------------------------------------
# Funções auxiliares para análise com nova lógica de contagem
# -------------------------------------------------------

MAPA_REGIOES_BRASIL = {
    "df": "centro-oeste", "go": "centro-oeste", "mt": "centro-oeste", "ms": "centro-oeste",
    "al": "nordeste", "ba": "nordeste", "ce": "nordeste", "ma": "nordeste", "pb": "nordeste",
    "pe": "nordeste", "pi": "nordeste", "rn": "nordeste", "se": "nordeste",
    "ac": "norte", "am": "norte", "ap": "norte", "pa": "norte", "ro": "norte", "rr": "norte", "to": "norte",
    "es": "sudeste", "mg": "sudeste", "rj": "sudeste", "sp": "sudeste",
    "pr": "sul", "rs": "sul", "sc": "sul"
}

def contar_registros_por_coluna(df, coluna, VarNome="VarAnalise"):
    """
    Conta ocorrências de registros (frequência) para representar a quantidade de bolsistas.
    """
    if coluna not in df.columns:
        print(f"Coluna '{coluna}' não encontrada.")
        return pd.DataFrame()

    resultado = (
        df[coluna]
        .dropna()
        .astype(str)
        .str.strip()
        .value_counts()
        .reset_index()
    )
    resultado.columns = [coluna, "quantidade"]

    # Armazena em variável global seguindo o padrão solicitado
    globals()[f"{VarNome}[PosicaoX]"] = resultado

    return resultado

def gerar_grafico_barras(df, coluna_x, titulo):
    if df.empty:
        return

    fig = px.bar(
        df,
        x=coluna_x,
        y="quantidade",
        title=titulo,
        text="quantidade"
    )
    fig.show()
```

```python
def executar_analise_textual(df_bolsistas=None, abas_carregadas=None):
    """
    Gera insights analíticos formatados em Markdown sobre Institutos e Programas.
    Itera sobre o Top 5 de IES Líderes detalhando a conexão de bolsistas.
    """

    # Verificação de segurança dos dados
    if df_bolsistas is None or abas_carregadas is None or "programas_fomento" not in abas_carregadas:
        print("Dados insuficientes para gerar a análise.")
        return

    df_b = df_bolsistas.copy()
    df_p = abas_carregadas["programas_fomento"].copy()

    # --- 1. SEÇÃO DE INSTITUTOS LÍDERES ---
    if 'instituicao' in df_p.columns:
        ies_lideres = df_p['instituicao'].value_counts().head(5)

        if not ies_lideres.empty:
            display(mk("## <font color='#73b2ff'>Análise Detalhada dos Institutos Líderes (Top 5)</font>"))

            # FOR: Iterando sobre cada IES no Top 5
            for rank, (ies_nome, qtd_projetos) in enumerate(ies_lideres.items(), 1):

                # Filtragem de conexões para a IES atual
                ids_projetos = df_p[df_p['instituicao'] == ies_nome]['id'].unique()
                df_conexoes = df_b[df_b['id'].isin(ids_projetos)].copy()
                total_conexoes = len(df_conexoes)

                # Cálculo de Militares vs Civis
                militares, civis = 0, 0
                if total_conexoes > 0 and 'tipo_ies' in df_conexoes.columns:
                    df_conexoes['tipo_ies'] = df_conexoes['tipo_ies'].astype(str).str.strip().str.upper()
                    tags_militares = ['EB', 'MB', 'FAB', 'MD'] #Não existe EGN como tag

                    militares = len(df_conexoes[df_conexoes['tipo_ies'].isin(tags_militares)])
                    civis = total_conexoes - militares

                # Montagem do texto formatado com f-strings
                texto_markdown = f"""
### {rank}º Lugar: **{ies_nome}**
* **Projetos Liderados:** {qtd_projetos}
* **Total de Conexões:** {total_conexoes} bolsistas vinculados.
* **Perfil dos Participantes:**
  * Militares: `{militares}`
  * Civis: `{civis}`

> A análise verificou que o **{ies_nome}** lidera projetos que conectam um total de **{total_conexoes}** bolsistas, fortalecendo a integração entre os setores.
"""
                display(mk(texto_markdown))

    # --- 2. SEÇÃO DE PROGRAMAS COM MAIS BOLSISTAS ---
    if 'qntd_de_bolsistas' in df_p.columns:
        prog_bolsistas = df_p.groupby('programa_de_fomento')['qntd_de_bolsistas'].sum().sort_values(ascending=False)

        if not prog_bolsistas.empty:
            display(mk("### <font color='#8fcf9b'>Programas com mais Bolsistas (Top 3)</font>"))
            for prog, qtd in prog_bolsistas.items():
                display(mk(f"* **{prog}**: {int(qtd)} bolsistas no total"))
```

```python
def executar_analise_textual_rede(df_bolsistas=None, abas_carregadas=None):
    """
    Gera insights analíticos avançados formatados em Markdown.
    Analisa a composição de participantes (Civis vs Militares) para cada IES líder.
    """

    if abas_carregadas is None or "programas_fomento" not in abas_carregadas:
        print("Dados insuficientes para gerar a análise.")
        return

    df_p = abas_carregadas["programas_fomento"].copy()

    if df_p is None or df_p.empty:
        print("Base de programas vazia.")
        return

    # 1. Normalização
    for col in ["programa_de_fomento", "instituicao", "tipo_ies", "ies_extenso"]:
        if col in df_p.columns:
            df_p[col] = df_p[col].astype(str).str.strip().str.upper()

    if "id" not in df_p.columns:
        df_p["id"] = df_p.index.astype(str)
    else:
        df_p["id"] = df_p["id"].astype(str).str.strip()

    df_p["qntd_de_bolsistas"] = pd.to_numeric(df_p.get("qntd_de_bolsistas", 0), errors="coerce").fillna(0)

    # 2. Processamento de Pesos e cruzamento com Bolsistas
    df_b = None
    if df_bolsistas is not None and not df_bolsistas.empty and "id" in df_bolsistas.columns:
        df_b = df_bolsistas.copy()
        df_b["id"] = df_b["id"].astype(str).str.strip()
        if "tipo_ies" in df_b.columns:
            df_b["tipo_ies"] = df_b["tipo_ies"].astype(str).str.strip().str.upper()

        contagem_bolsistas = df_b.groupby("id").size().reset_index(name="bolsistas_detalhado")
        df_p = df_p.merge(contagem_bolsistas, on="id", how="left")
        df_p["bolsistas_detalhado"] = df_p["bolsistas_detalhado"].fillna(0)
        df_p["peso_final"] = df_p[["qntd_de_bolsistas", "bolsistas_detalhado"]].max(axis=1)
    else:
        df_p["peso_final"] = df_p["qntd_de_bolsistas"]

    # 3. Análise de Liderança Institucional
    ies_lideres = df_p.groupby("instituicao").agg(
        qtd_projetos=("id", "nunique"),
        total_bolsistas=("peso_final", "sum"),
        tipo_ies_lider=("tipo_ies", "first"),
        ies_nome_extenso=("ies_extenso", "first")
    ).sort_values(by=["qtd_projetos", "total_bolsistas"], ascending=False).head(5)

    display(mk("# <font color='#73b2ff'>Relatório de Inteligência Institucional</font>"))
    display(mk("## <font color='#73b2ff'>1. Top 5 Instituições Ícones (Liderança de Projetos)</font>"))

    tags_militares = ["EB", "MB", "FAB", "MD", "MILITAR"]

    for rank, (ies_sigla, row) in enumerate(ies_lideres.iterrows(), 1):
        nome_exibicao = row['ies_nome_extenso'] if pd.notna(row['ies_nome_extenso']) and row['ies_nome_extenso'] != "NAN" else ies_sigla

        # Buscar todos os IDs de projetos dessa IES líder
        ids_da_ies = df_p[df_p["instituicao"] == ies_sigla]["id"].unique()

        militares_count = 0
        civis_count = 0

        if df_b is not None:
            # Filtrar bolsistas vinculados a esses projetos
            participantes = df_b[df_b["id"].isin(ids_da_ies)]
            if not participantes.empty and "tipo_ies" in participantes.columns:
                militares_count = len(participantes[participantes["tipo_ies"].isin(tags_militares)])
                civis_count = len(participantes) - militares_count

        total_conexoes = militares_count + civis_count
        percent_mil = (militares_count / total_conexoes * 100) if total_conexoes > 0 else (100 if row['tipo_ies_lider'] in tags_militares else 0)

        texto_markdown = f"""
### **{rank}º - {nome_exibicao} ({ies_sigla})**
* **Status Operacional:** {int(row['qtd_projetos'])} Projetos Ativos
* **Força de Pesquisa:** {int(row['total_bolsistas'])} Bolsistas Conectados
* **Natureza da Líder:** {'Militar' if row['tipo_ies_lider'] in tags_militares else 'Civil'}
* **Mix de Participantes (Conexões):** `{percent_mil:.1f}%` Militares vs `{100-percent_mil:.1f}%` Civis.

> **Insight:** A {ies_sigla} lidera uma rede com **{total_conexoes}** participantes identificados, demonstrando sua capacidade de integração multissetorial.
"""
        display(mk(texto_markdown))
```

```python
# -------------------------------------------------------
# 3.5. Funções de Execução Modular (Gráficos e Redes)
# -------------------------------------------------------

def executar_analise_bolsistas_ies(df_bolsistas):
    if df_bolsistas is not None:
        print("Gerando análise de Bolsistas por IES...")
        grafico_bolsistas_por_ies(df_bolsistas)
    else:
        print("Dados de bolsistas não disponíveis.")

def executar_analise_regiao_nacional(df_bolsistas):
    if df_bolsistas is not None:
        print("Gerando análise de Distribuição por Região...")
        grafico_ies_por_regiao_nacional(df_bolsistas)

def executar_analise_tipo_ies(df_bolsistas):
    if df_bolsistas is not None:
        if "tipo_ies" not in df_bolsistas.columns and "sigla_ies" in df_bolsistas.columns:
            df_bolsistas["tipo_ies"] = df_bolsistas["sigla_ies"].astype(str).str.strip().str.upper().apply(
                lambda x: "MILITAR" if x in ["EB", "MB", "FAB", "MD"] else "CIVIL"
            )

        if "tipo_ies" in df_bolsistas.columns:
            grafico_bolsistas_por_tipo_ies(df_bolsistas)
            print("\n")
            grafico_bolsistas_ies_militar_civil(df_bolsistas)

def executar_analise_internacional(df_programas):
    if df_programas is not None:
        print("Gerando análise Internacional...")
        grafico_ies_por_regiao_internacional(df_programas)

def executar_analise_programas_fomento(df_programas):
    if df_programas is not None:
        print("Gerando análise de Programas de Fomento...")
        grafico_ies_por_programa_fomento(df_programas)

def executar_todas_as_redes(df_programas, df_bolsistas, ocultar_programas=None):
    if df_programas is not None:
        print("Gerando Redes de Conexão (Network)...\n")

        print("1. Rede Profissional (Apresentações)")
        # Repassa o filtro para a função específica
        gerar_rede_ies_programas_profissional(df_programas, df_bolsistas, ocultar_programas=ocultar_programas)
        print("\n")

        print("2. Conexão Parceiro -> Programa -> IES")
        gerar_rede_parceiros_prog_ies_profissional(df_programas)
        print("\n")

        print("3. Rede Nacional (Filtro Brasil)")
        gerar_rede_nacional(df_programas)
        print("\n")

        print("4. Rede Internacional (Filtro Exterior)")
        gerar_rede_internacional(df_programas)
        print("\n")

        if df_bolsistas is not None:
            print("5. Rede de Conexão IES Líder <-> Bolsistas")
            gerar_rede_ies_lid_bols(df_programas, df_bolsistas)
```
